#### 기본 구성

```python
import matplotlib.pyplot as plt

first_x_data = [1,2,3,4]
first_y_data = [1,2,3,4] # x,y 갯수 맞아야 됨
second_x_data = [2,3,4,1]
second_y_data = [1,2,3,4] # x,y 중 하나를 sorting 해야 됨?
third_x_data = [2,3,4,1]
third_y_data = [1,2,3,4]
fourth_x_data = [4,3,2,1]
fourth_y_data = [4,3,2,1]
```

### 단변수 표현

```python
plt.plot(third_x_data)
plt.show()
```

### 다변수 표현

```python
plt.plot(first_x_data, first_y_data)
plt.plot(third_x_data, third_y_data)
plt.show()
```

### 설명 구성 

```python

plt.plot(first_x_data, first_y_data, label='first')
plt.plot(third_x_data, third_y_data, label='third')
plt.title('Visulliazation with title')
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.legend()
plt.show()
```

### 한글 처리 및 마이너스 기호 깨짐 방지

```python
import matplotlib as mpl # 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
```

```python

plt.plot(first_x_data, first_y_data, label='first 범례')
plt.plot(third_x_data, third_y_data, label='third 범례')
plt.title('Visulliazation with title 이름')
plt.xlabel('x_data x축')
plt.ylabel('y_data y축')
plt.legend()
plt.savefig('visullizations_plots.png')
plt.show()
```

### figure 넓이, 높이 조정 

```python
# plt.figure(figsize=(10,6)) # inch (10 inch)(w) * (6 inch)(h)
plt.figure(figsize=(10,6), dpi=100) # 

plt.plot(first_x_data, first_y_data, label='first 범례')
plt.plot(third_x_data, third_y_data, label='third 범례')
plt.title('Visulliazation with title 이름')
plt.xlabel('x_data x축')
plt.ylabel('y_data y축')
plt.legend()
plt.savefig('visullizations_plots.png')
plt.show()
```
