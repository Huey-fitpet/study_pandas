## visuallization

### matplotlib

```python
import matplotlib.pyplot as plt
```

#### 기본 구성

```python
first_x_data = [1,2,3,4]
first_y_data = [1,2,3,4] # x,y 갯수 맞아야 됨
second_x_data = [2,3,4,1]
second_y_data = [1,2,3,4] # x,y 중 하나를 sorting 해야 됨?
third_x_data = [2,3,4,1]
third_y_data = [1,2,3,4]
fourth_x_data = [4,3,2,1]
fourth_y_data = [4,3,2,1]
```

```python
plt.plot(first_x_data, first_y_data)
plt.plot(third_x_data, third_y_data)
plt.show()
```

### 정렬
- add_subplot(), subplots()

### 구성 요소 추가

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(1,2,figsize=(14,10)) # 외곽 사이즈

fig.suptitle('Super Title')
# fig.supxlabel('super x')
# fig.supylabel('super y')

axes[0].plot(first_x_data, first_y_data, label='first' )
axes[0].plot(second_x_data, second_y_data, label='second')
axes[0].set_title('axes first')
axes[0].set_xlabel('x_data')
axes[0].legend()
# axes[0].set_ylabel('y_data')

axes[1].plot(third_x_data, third_y_data, label='third')
axes[1].set_title('axes third')
# axes[1].set_xlabel('x_data')
axes[1].set_ylabel('y_data')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(1,4,figsize=(14,10)) # 외곽 사이즈

axes[0].plot(first_x_data, first_y_data)
axes[1].plot(third_x_data, third_y_data)
axes[2].plot(second_x_data, second_y_data)
axes[3].plot(fourth_x_data, fourth_y_data)

plt.show()
```

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(2,4,figsize=(14,10)) # 외곽 사이즈

# axes[0][1].plot(first_x_data, first_y_data)
# axes[1][0].plot(third_x_data, third_y_data)
# axes[0][2].plot(second_x_data, second_y_data)
# axes[0][3].plot(fourth_x_data, fourth_y_data)

axes[0,1].plot(first_x_data, first_y_data, label='first')
axes[0,1].plot(third_x_data, third_y_data, label='third')
# axes[0,1].title('Visulliazation with title')
# axes[0,1].xlabel('x_data')
# axes[0,1].ylabel('y_data')
axes[0,1].legend()
axes[1,0].plot(third_x_data, third_y_data)
axes[0,2].plot(second_x_data, second_y_data)
axes[0,3].plot(fourth_x_data, fourth_y_data)

plt.show()
```

### figure size 조정

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(1,2,figsize=(14,10),dpi=100) # 외곽 사이즈

fig.suptitle('Super Title')
# fig.supxlabel('super x')
# fig.supylabel('super y')

axes[0].plot(first_x_data, first_y_data, label='first' )
axes[0].plot(second_x_data, second_y_data, label='second')
axes[0].set_title('axes first')
axes[0].set_xlabel('x_data')
axes[0].legend()
# axes[0].set_ylabel('y_data')

axes[1].plot(third_x_data, third_y_data, label='third')
axes[1].set_title('axes third')
# axes[1].set_xlabel('x_data')
axes[1].set_ylabel('y_data')
axes[1].legend()

plt.tight_layout()
plt.show()
```
