## Load from DB

```python
from pymongo import MongoClient
# 네트워크 연결 규칙 : protocol://ip:port/path 
client = MongoClient('mongodb://192.168.0.91:27017/')

db_name = client["DB_PANDAS"]
phone_collection = db_name["COL_PHONE_LIST"]
```

```python
find_data = phone_collection.find()
```

```python
phone_data = list(find_data)
phone_data
```

```python
import pandas as pd 
```

```python
df_phone_list = pd.DataFrame(phone_data)
```

```python
df_phone_list.info()
```

```python
df_phone_list.describe(include="all")
```

## Save to DB

```python
df_phone_list_without_id = df_phone_list.drop(labels='_id',axis=1)
df_phone_list_without_id
```

```python
data_to_insert = df_phone_list_without_id.to_dict(orient="records")
data_to_insert
```

```python
phone_collection.insert_many(data_to_insert)
```
