## 코스피 정보 출력

```python
import matplotlib.pyplot as plt
import matplotlib as mpl # 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# data => dataframe => matplot
# api 
import FinanceDataReader as fdr

kospi = fdr.DataReader('KS11','2023') # KS000011

kospi.head()
```

```python
type(kospi)
```

```python
kospi.info()
kospi.describe()
```

### plot 그리기

```python
plt.figure(figsize=(10,6), dpi=100)

plt.plot(kospi['Low'], label='Low')
plt.plot(kospi['High'], label='High')
plt.title('코스피 지수 (2023~)')
plt.xlabel('날짜')
plt.ylabel('가격')
plt.legend()
plt.show()
```

```python
plt.plot(kospi['Amount'], label='Amount') # 총 양 인가봄 
plt.plot(kospi['Volume'], label='Volume') 
plt.title('코스피 지수 (2023~)')
plt.xlabel('날짜')
plt.ylabel('가격')
plt.legend()
plt.show()
```
