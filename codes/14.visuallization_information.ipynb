### 지수화(indexing)
- 서로 다른 값차이가 많이 날때 추세만 같이 보기 위해 처리하는 것

```python
import numpy as np
```

```python
first_array = np.array([10,20,30])
second_array = np.array([1000,2000,3000]) # 백분위 처리하면 비교하기 쉬움
```

```python
(first_array / first_array[0]) * 100
```

```python
(second_array / second_array[0]) * 100
```

```python
tway = fdr.DataReader(symbol='091810', start='2021', end='2022')
tway.head()
```

```python
samsung = fdr.DataReader(symbol='005930', start='2021', end='2022')
samsung.head()
```

```python
tway.shape, samsung.shape
```

```python
plt.plot(tway.index, tway['Close'], label='tway Close') # 총 양 인가봄 
plt.plot(samsung.index, samsung['Close'], label='samsung Close') 
plt.title('티웨이 삼성 종가 비교 (2021~2022)')
plt.xlabel('날짜')
plt.ylabel('가격')
plt.legend()
plt.show()
```

```python
tway_close = (tway['Close'] / tway['Close'].iloc[0]) * 100
samsung_close = (samsung['Close'] / samsung['Close'].iloc[0]) * 100
```

```python
plt.plot(tway_close.index, tway_close, label='tway Close') # 총 양 인가봄 
plt.plot(samsung_close.index, samsung_close, label='samsung Close') 
plt.title('티웨이 삼성 종가 비교 (2021~2022)')
plt.xlabel('날짜')
plt.ylabel('지수')
plt.legend()
plt.show()
```

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(1,2,figsize=(14,10),dpi=100) # 외곽 사이즈

fig.suptitle('Super Title')
# fig.supxlabel('super x')
# fig.supylabel('super y')

axes[0].plot(tway.index, tway['Close'], label='tway Close') # 총 양 인가봄
axes[0].plot(samsung.index, samsung['Close'], label='samsung Close') 
axes[0].set_title('axes first')
axes[0].set_xlabel('x_data')
axes[0].legend()
# axes[0].set_ylabel('y_data')

axes[1].plot(tway_close.index, tway_close, label='tway Close') # 총 양 인가봄
axes[1].plot(samsung_close.index, samsung_close, label='samsung Close') 
axes[1].set_title('axes third')
# axes[1].set_xlabel('x_data')
axes[1].set_ylabel('y_data')
axes[1].legend()

plt.tight_layout()
plt.show()
```
