## Load data 

```python
from pymongo import MongoClient
# 네트워크 연결 규칙 : protocol://ip:port/path 
client = MongoClient('mongodb://192.168.0.50:27017/')

db_name = client["DB_SGMN"]
col_src_name = db_name["COL_KOSIS_BUSINESS_CYCLE_SURVEY_FORECAST"]
```

```python
data_cursor = col_src_name.find({
    "PRD_DE": {
        "$gte": "201501"  # 2015년 1월 이상
        },
    "C1_NM" : "서비스업"  # 추가 조건
})
data_list = list(data_cursor)
print(len(data_list))
data_list
```

```python
data_cursor = col_src_name.find({
    "$and": [
        {
            "PRD_DE": {
                "$gte": "201501"  # 2015년 1월 이상
            }
        },
        {
            "C1_NM" : {
               "$eq" : "서비스업"  # 추가 조건
            }
        },
        {
            "C2_NM_ENG": {
                "$regex": "growth",  
                "$options": "i"         
            }
        }
    ]
})
data_list = list(data_cursor)
print(len(data_list))
data_list
```
