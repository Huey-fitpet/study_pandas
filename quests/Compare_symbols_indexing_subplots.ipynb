```python
import matplotlib.pyplot as plt
import matplotlib as mpl # 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# api 
import FinanceDataReader as fdr

kospi = fdr.DataReader(symbol='KS11', start='1990') # KS000011
ssec = fdr.DataReader(symbol='SSEC', start='1990') # 상해 종합지수 Shanghai (SSEC -Shanghai Composite)
hsi = fdr.DataReader(symbol='HSI', start='1990') # 항셍지수 (HSI - Hang Seng)
n255 = fdr.DataReader(symbol='N225', start='1990') # 일본 닛케이지수 (N225 - Nikkei 225)
ftse = fdr.DataReader(symbol='FTSE', start='1990') # 영국 FTSE100 (FTSE 100 - Financial Times Stock Exchange)
fchi = fdr.DataReader(symbol='FCHI', start='1990') # 프랑스 FCHI 지수 (CAC 40 - CAC quarante)
gdaxi = fdr.DataReader(symbol='GDAXI', start='1990') # 독일 닥스지수  (DAX30 - germany-30)
```

```python
kospi.info()
ssec.info()
hsi.info()
n255.info()
ftse.info()
fchi.info()
gdaxi.info()
```

```python
kospi_close = (kospi['Close'] / kospi['Close'].iloc[0]) * 100
ssec_close = (ssec['Close'] / ssec['Close'].iloc[0]) * 100
hsi_close = (hsi['Close'] / hsi['Close'].iloc[0]) * 100
n255_close = (n255['Close'] / n255['Close'].iloc[0]) * 100
ftse_close = (ftse['Close'] / ftse['Close'].iloc[0]) * 100
fchi_close = (fchi['Close'] / fchi['Close'].iloc[0]) * 100
gdaxi_close = (gdaxi['Close'] / gdaxi['Close'].iloc[0]) * 100

global_list = [
    kospi_close,
    ssec_close, 
    hsi_close,
    n255_close, 
    ftse_close, 
    fchi_close ,
    gdaxi_close
]
```

```python

plt.figure(figsize=(14,10), dpi=100)
plt.plot(kospi_close.index, kospi_close, label='kospi ') 
plt.plot(ssec_close.index, ssec_close, label='ssec ') 
plt.plot(hsi_close.index, hsi_close, label='hsi ') 
plt.plot(n255_close.index, n255_close, label='n255 ') 
plt.plot(ftse_close.index, ftse_close, label='ftse ') 
plt.plot(fchi_close.index, fchi_close, label='fchi ') 
plt.plot(gdaxi_close.index, gdaxi_close, label='gdaxi ') 
plt.title('국가 지수별 종가 비교 (1990~)')
plt.xlabel('날짜')
plt.ylabel('지수')
plt.legend()
plt.show()
```

```python
global_name_list = [
    "kospi",
    "ssec",
    "hsi",
    "n255",
    "ftse",
    "fchi",
    "gdaxi",
]
```

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈

fig, axes = plt.subplots(len(global_list)-1,1,figsize=(10,20), dpi=100) # 외곽 사이즈

fig.suptitle('국가별 kospi와 종가 비교 (1990~)')
# fig.supxlabel('super x')
# fig.supylabel('super y')
kospi_close = global_list[0]
for num, (global_index, global_name) in enumerate(zip(global_list[1:], global_name_list[1:])):
    axes[num].plot(kospi_close.index, kospi_close, label='kospi' )
    axes[num].plot(global_index.index, global_index, label=global_name)
    # axes[num].set_title('axes first')
    axes[num].set_xlabel('년도')
    axes[num].set_ylabel('지수')
    axes[num].legend()


# plt.tight_layout()
plt.show()
```
