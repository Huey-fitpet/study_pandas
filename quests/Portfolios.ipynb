```python
# 상관계수 매트릭스를 일렬로 펼치기
def get_correlation_pairs(correlation_matrix):
    # 상관계수 매트릭스를 DataFrame으로 변환 (이미 DataFrame인 경우 생략 가능)
    corr_df = pd.DataFrame(correlation_matrix)
    
    # 상관계수 쌍을 리스트로 변환
    pairs = []
    for i in corr_df.index:
        for j in corr_df.columns:
            if i < j:  # 중복 제거 (대각선 위쪽만 사용)
                pairs.append({
                    'pair': f'{i} - {j}',
                    'correlation': corr_df.loc[i, j]
                })
    
    # DataFrame으로 변환
    pairs_df = pd.DataFrame(pairs)
    
    # 양의 상관관계 상위 10개
    positive_corr = pairs_df[pairs_df['correlation'] > 0].nlargest(10, 'correlation')
    
    # 음의 상관관계 상위 10개 (절대값이 가장 큰 음수)
    negative_corr = pairs_df[pairs_df['correlation'] < 0].nsmallest(10, 'correlation')
    
    # 절대값 기준 하위 10개 (상관관계가 가장 약한 쌍)
    lowest_corr = pairs_df.iloc[pairs_df['correlation'].abs().argsort()][:10]
    
    return positive_corr, negative_corr, lowest_corr

# 결과 출력
positive_corr, negative_corr, lowest_corr = get_correlation_pairs(correlaiton_matrix)
positive_corr
```

2. 샤프 지수를 사용한 최적 포트폴리오 구성
목표: 상관계수 분석을 통해 선별한 일부 ETF 종목으로 최적의 투자 비율을 찾습니다.
요구 사항:
선택된 ETF 종목의 일간 수익률 데이터를 기반으로 평균 수익률과 공분산 행렬을 계산하세요.
무위험 이자율을 설정하고 샤프 지수를 최대화하는 포트폴리오의 가중치를 찾으세요.
최적 가중치, 포트폴리오 기대 수익률, 변동성, 샤프 지수를 보고하세요.

```python
# start_date = f'20220101'
# end_date = f'20231231'

stock_df = pd.DataFrame()
for symbol in tickers:
    stock_df[symbol] = stock.get_market_ohlcv_by_date(start_date, end_date, symbol)['종가'] # 네이버

stock_df.head()
```

```python
daily_return = stock_df.pct_change(periods=1) # shift and percent 까지 해서 수익률 나옴.
annual_return = daily_return.mean() * 252
daily_cov = daily_return.cov()
annual_cov = daily_cov * 252
```

```python
import numpy as np
```

```python
# 여러 포트폴리오 비율 적용
fiture_count = 10000
portfolio_return_of_rate = []
portfolio_risk = []
portfolio_weights = []
sharpe_ration = [] 

for num in range(fiture_count):
    weights = np.random.random(len(tickers))
    weights = weights / np.sum(weights)
    return_of_rate = np.dot(weights, annual_return) # y
    risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights))) # x

    portfolio_return_of_rate.append(return_of_rate)
    portfolio_risk.append(risk)
    portfolio_weights.append(weights)
    sharpe_ration.append(return_of_rate / risk)

```

```python
# Convert Dataframe : Return of rate, Rist, weights
portfolio = {
    "Return_of_rate" : portfolio_return_of_rate,
    "Risk" : portfolio_risk,
    "Sharpe": sharpe_ration
}

for idx, symbol in enumerate(tickers):
    portfolio[symbol] =  [weight[idx] for weight in portfolio_weights]


df_portfolio = pd.DataFrame(portfolio)
df_portfolio.head()
```

```python
maxindex = df_portfolio['Sharpe'].idxmax()
df_portfolio.iloc[maxindex] # 가장 좋은 포트폴리오 
```

3. 선별 종목 별 투자 전략 적용
목표: 샤프 지수로 최적화된 포트폴리오 내 각 ETF 종목에 대해 적절한 투자 전략을 선택합니다.
요구 사항:
할로윈 투자 전략: 특정 종목에 대해 매년 10월~4월에만 매수 전략을 적용하고 나머지 기간에는 매도.
변동성 돌파 전략: 특정 종목에 대해 전날의 고점과 저점 차이를 기준으로 돌파 매수 신호를 탐지.
PBR+PER 콤보 전략: PBR과 PER 값이 낮은 종목 위주로 선별하여 투자. 데이터를 기반으로 최적의 진입 타이밍 탐색.
RSI 전략: 상대강도지수(RSI)가 과매도 구간(30 이하)일 때 매수하고 과매수 구간(70 이상)일 때 매도.
볼린저 밴드 전략: 볼린저 밴드 하단에 닿을 때 매수하고 상단에 닿을 때 매도.
결과 제출:
각 종목에 대한 전략 선정 이유를 작성하세요.
전략을 백테스트하고 수익률을 비교하세요.

```python
def calc_halloween_rate_of_return(df):
    # 결과를 저장할 새로운 컬럼 생성
    df['Halloween_Return'] = float('nan')  # 모든 값을 우선 NaN으로 초기화
    
    cum_rate_of_return = 1
    start_year = df.index[0].year
    end_year = df.index[-1].year
    
    for year in range(start_year, end_year):
        # 매수 시점 (11월)
        buy_date = f'{year}-11'
        # 매도 시점 (다음해 4월)
        sell_date = f'{year+1}-04'
        
        try:
            # 해당 기간의 시작 가격
            buy_price = df.loc[buy_date].iloc[0]['시가']
            
            # 11월부터 다음해 4월까지의 기간 선택
            mask = (df.index >= buy_date) & (df.index <= f'{year+1}-04-30')
            
            # 해당 기간 동안의 일별 수익률 계산
            df.loc[mask, 'Halloween_Return'] = (df.loc[mask, '종가'] / buy_price) * cum_rate_of_return
            
            # 매도 시점의 가격으로 누적 수익률 업데이트
            sell_price = df.loc[sell_date].iloc[-1]['종가']
            cum_rate_of_return *= (sell_price / buy_price)
            
        except KeyError:
            continue
    
    # CAGR 계산
    diff_date = df.index[-1] - df.index[0]
    df['Halloween_CAGR'] = ((df['Halloween_Return'] ** (1/(diff_date.days/365)))-1) * 100
    
    return df
```

```python
def calc(series):
    result = []
    current_state = False  # 초기 상태는 False
    
    for value in series:
        if current_state:  # True 상태일 때
            if value >= 70:  # 70 이상이면 False로 전환
                current_state = False
        else:  # False 상태일 때
            if value <= 30:  # 30 이하면 True로 전환
                current_state = True
        
        result.append(current_state)
    
    return result
```

```python
# RSI
def Calc_RSI_CAGR(stock_df_close): 
    stock_df_close['변화량'] = stock_df_close['종가'] - stock_df_close['종가'].shift(1)
    condition_up = stock_df_close['변화량'] >= 0 
    condition_down = stock_df_close['변화량'] < 0 
    stock_df_close.loc[condition_up, '상승폭'] = stock_df_close['변화량']
    stock_df_close.loc[condition_down, '하락폭'] = -stock_df_close['변화량']
    stock_df_close = stock_df_close.fillna(0)
    stock_df_close['AU'] = stock_df_close['상승폭'].rolling(window=14).mean()
    stock_df_close['DU'] = stock_df_close['하락폭'].rolling(window=14).mean()
    stock_df_close['RSI'] = (stock_df_close['AU'] / (stock_df_close['AU'] + stock_df_close['DU'])) * 100 
    stock_df_close['bool'] = calc(stock_df_close['RSI'])
    stock_df_close['전일종가'] = stock_df_close['종가'].shift(periods=1)
    delta_days = stock_df_close.index[-1] - stock_df_close.index[0]
    year = delta_days.days / 365

    stock_df_close['수익률'] = np.where(stock_df_close['bool'], stock_df_close['종가'] / stock_df_close['전일종가'], 1)
    stock_df_close['누적수익률'] = stock_df_close['수익률'].cumprod()
    CAGR = stock_df_close['누적수익률'].iloc[-1] ** (1/year) -1
    return CAGR * 100
```

```python
def calc_band(closes, highs, lows):
    result = []
    current_state = False  # 초기 상태는 False
    
    for close, high, low in zip(closes, highs, lows):
        if current_state:  # True 상태일 때
            if close >= high:  # 70 이상이면 False로 전환
                current_state = False
        else:  # False 상태일 때
            if close <= low:  # 30 이하면 True로 전환
                current_state = True
        
        result.append(current_state)
    
    return result
```

```python
# bollinger band
def Calc_Bollinger_CAGR(stock_df_close): 
    stock_df_close['중심선'] = stock_df_close['종가'].rolling(window=14).mean()
    K = 2
    std_deviation = stock_df_close['종가'].rolling(window=14).std()
    stock_df_close['상단선'] = stock_df_close['중심선'] + K * std_deviation
    stock_df_close['하단선'] = stock_df_close['중심선'] - K * std_deviation
    stock_df_close['bool'] = calc_band(stock_df_close['종가'],stock_df_close['상단선'],stock_df_close['하단선'])
    stock_df_close['전일종가'] = stock_df_close['종가'].shift(periods=1)
    delta_days = stock_df_close.index[-1] - stock_df_close.index[0]
    year = delta_days.days / 365

    stock_df_close['수익률'] = np.where(stock_df_close['bool'], stock_df_close['종가'] / stock_df_close['전일종가'], 1)
    stock_df_close['누적수익률'] = stock_df_close['수익률'].cumprod()
    CAGR = stock_df_close['누적수익률'].iloc[-1] ** (1/year) -1
    return CAGR * 100
```

```python
# 변동폭
def Calc_Volatility_CAGR(stock_df_close): 
    stock_df_close['변동폭'] = stock_df_close['고가'] - stock_df_close['저가']
    stock_df_close['전일변동폭'] = stock_df_close['변동폭'].shift(periods=1) #전일 변동폭이라서 
    K = 1/2
    stock_df_close['목표가'] = stock_df_close['시가'] + (stock_df_close['전일변동폭'] * K)
    conditions = stock_df_close['고가'] >= stock_df_close['목표가']  # 매도 조건
    stock_df_close['수익률'] = np.where(conditions, stock_df_close['종가'] / stock_df_close['목표가'], 1)
    delta_days = stock_df_close.index[-1] - stock_df_close.index[0]
    year = delta_days.days / 365
    stock_df_close['누적수익률'] = stock_df_close['수익률'].cumprod()
    CAGR = stock_df_close['누적수익률'].iloc[-1] ** (1/year) -1
    return CAGR * 100
```

```python
# tickers 만큼 돌면서 각 누적수익률 구하는 함수 통해서 값 출력해서 비교하게 해보기 
# ticker index 

# 결과를 저장할 딕셔너리
results = {ticker: {} for ticker in tickers}

# 각 ticker에 대해 전략 실행
for ticker in tickers:
    # 주가 데이터 가져오기
    df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)
    
    # 각 전략별 수익률 계산 및 저장
    # results[ticker]['Halloween'] = calc_holloween(df)
    results[ticker]['RSI'] = Calc_RSI_CAGR(df)
    results[ticker]['Bollinger'] = Calc_Bollinger_CAGR(df)
    results[ticker]['Volatility'] = Calc_Volatility_CAGR(df)
    temp_df = calc_halloween_rate_of_return(df)
    temp_df = temp_df.dropna(subset=['Halloween_CAGR'])
    results[ticker]['Halloween'] = temp_df['Halloween_CAGR'].iloc[-1]

# 딕셔너리를 DataFrame으로 변환
result_df = pd.DataFrame.from_dict(results, orient='index')
result_df
```
