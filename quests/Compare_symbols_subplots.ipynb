```python
import matplotlib.pyplot as plt
import matplotlib as mpl # 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# api 
import FinanceDataReader as fdr
```

```python
kospi = fdr.DataReader('KS11','2023') # KS000011
kospi.head()
```

```python
kosdaq = fdr.DataReader('KQ11','2023') # KS000011
kosdaq.head()
```

```python
# fig, axes = plt.subplots(2,2,figsize=(14,10)) # 외곽 사이즈
fig, axes = plt.subplots(4,1,figsize=(10,14),dpi=100) # 외곽 사이즈

fig.suptitle('코스피 코스닥 비교 (2023~)')
# fig.supxlabel('super x')
# fig.supylabel('super y')

axes[0].plot(kospi['Close'], label='kospi Close' )
axes[0].plot(kosdaq['Close'], label='kosdaq Close')
axes[0].set_title('종가 비교')
axes[0].set_xlabel('년도')
axes[0].set_ylabel('가격')
axes[0].legend()

axes[1].plot(kospi['Volume'], label='kospi Volume' )
axes[1].plot(kosdaq['Volume'], label='kosdaq Volume')
axes[1].set_title('거래량 비교')
axes[1].set_xlabel('년도')
axes[1].set_ylabel('가격')
axes[1].legend()

axes[2].plot(kospi['Amount'], label='kospi Amount' )
axes[2].plot(kosdaq['Amount'], label='kosdaq Amount')
axes[2].set_title('총량 비교')
axes[2].set_xlabel('년도')
axes[2].set_ylabel('가격')
axes[2].legend()

axes[3].plot(kospi['MarCap'], label='kospi MarCap' )
axes[3].plot(kosdaq['MarCap'], label='kosdaq MarCap')
axes[3].set_title('MarCap 비교')
axes[3].set_xlabel('년도')
axes[3].set_ylabel('가격')
axes[3].legend()

plt.tight_layout()
plt.show()
```
