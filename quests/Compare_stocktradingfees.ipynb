```python
import pandas as pd 
```

```python
file_path = f"../datasets/주식거래 수수료_20241217_50만원.xls"
df_xlsx = pd.read_excel(io=file_path, header=2)
df_xlsx.head()
```

```python
df_filter = df_xlsx[df_xlsx["구분"]=="변경후"]
df_sort = df_filter.sort_values(by=["스마트폰","스마트폰.1","HTS","HTS.1"],ascending=[True,True,True,True])
df_sort.head()
```

```python
df_sort_smartphone = df_filter.sort_values(by=["스마트폰"],ascending=[True])
df_sort_smartphone1 = df_filter.sort_values(by=["스마트폰.1"],ascending=[True])
df_sort_HTS = df_filter.sort_values(by=["HTS"],ascending=[True])
df_sort_HTS1 = df_filter.sort_values(by=["HTS.1"],ascending=[True])
print("증권사지점개설계좌 스마트폰 저렴한 TOP 5.")
print(df_sort_smartphone[['회사명', '스마트폰']].iloc[:5]) # "스마트폰"
print()
print("은행개설계좌 스마트폰 저렴한 TOP 5.")
print(df_sort_smartphone1[['회사명', '스마트폰.1']].iloc[:5]) # "스마트폰1"
print()
print("증권사지점개설계좌 HTS 저렴한 TOP 5.")
print(df_sort_HTS[['회사명', 'HTS']].iloc[:5]) # "HTS"
print()
print("은행개설계좌 HTS 저렴한 TOP 5.")
print(df_sort_HTS1[['회사명', 'HTS.1']].iloc[:5]) # "HTS1"
```
