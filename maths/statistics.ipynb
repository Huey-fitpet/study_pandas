## 평균, 중앙값과 첨도의 관계

```python

# 데이터 생성 
data_symmetric = [1, 2, 3, 4, 5, 6, 7, 8, 9] 

# 대칭 분포 
data_positive_skew = [1, 2, 2, 3, 3, 3, 10, 10, 50] # 오른쪽 꼬리 
data_negative_skew = [-50, -10, -10, -3, -3, -3, -2, -2, -1] # 왼쪽 꼬리
```

```python
import numpy as np
from scipy.stats import skew, kurtosis # 왜도 계산 함수
from numpy import ndarray as nparray
```

```python

def cal_stats(data: nparray) -> tuple[float, float, float, float]:
    mean = np.mean(data)
    median = np.median(data)
    kurt = kurtosis(data)  # kurt > 0 -> 오른쪽 꼬리
    skew_np = skew(data)    # skew > 0 -> 오른쪽 꼬리
    return mean, median, kurt, skew_np
```

```python
mean, median, kurt, skew_np = cal_stats(data_positive_skew)
mean, median, kurt, skew_np
```

```python
mean, median, kurt, skew_np = cal_stats(data_negative_skew)
mean, median, kurt, skew_np
```

```python
mean, median, kurt, skew_np = cal_stats(data_symmetric)
mean, median, kurt, skew_np
```
